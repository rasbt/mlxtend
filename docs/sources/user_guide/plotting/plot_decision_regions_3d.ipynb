# Plotting Decision Regions in 3D

The `plot_decision_regions_3d` function is an extension of the classic 2D decision region plot. It allows for the visualization of decision boundaries in a 3D space by "stacking" multiple 2D slices along a third axis.

from mlxtend.plotting import plot_decision_regions_3d

## Overview

Visualizing decision boundaries for models with more than two features is challenging. `plot_decision_regions_3d` solves this by taking "snapshots" of the decision regions at different levels of a third feature (Z-axis). This is particularly useful for:
* Understanding how a classifier's confidence or boundary shifts with a continuous third feature.
* Educational purposes to demonstrate high-dimensional decision surfaces.

## Example 1 - Stacked Slices for SVM

In this example, we train a Support Vector Machine on a 3-feature dataset and visualize the boundaries at 5 different Z-levels.

```python
import matplotlib.pyplot as plt
from mlxtend.plotting import plot_decision_regions_3d
from sklearn.svm import SVC
from sklearn.datasets import make_blobs
import numpy as np

# 1. Generate 3D data
X, y = make_blobs(n_samples=100, n_features=3, centers=2, random_state=42, cluster_std=1.5)

# 2. Train a classifier
clf = SVC(kernel='linear', C=1.0)
clf.fit(X, y)

# 3. Define the Z-levels (slices) we want to see
z_levels = np.linspace(X[:, 2].min(), X[:, 2].max(), num=5)

# 4. Plotting
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

plot_decision_regions_3d(X, y, clf=clf, z_slices=z_levels, ax=ax)

plt.title('3D Stacked Decision Regions')
plt.show()
```
